```julia
import TestImages: testimage
import ImageView: imshow

img = testimage("mandrill")
(xsize, ysize) = size(img)

function warp3(x, y, knob) # 0:31
    clip(n, nsize) = max(min(nsize, n), 1)
    (x0, y0) =
        [(clip(convert(Int, floor(x - 10 * (knob / 32) * sin(y/4))), xsize), y),
         (clip(convert(Int, floor(x + 10 * ((knob - 16) / 16) * sin(y/4))), xsize), y) # normalize to get +/-
        ][end]
    img[x0, y0]
end
```

```julia
import Images: Gray
img = 1*Gray.(img)
```

```julia
imgs3 = [[warp3(x, y, i)
         for x in 1:xsize, y in 1:ysize]
         for i in 0:31]

zsize = length(imgs3)
```

```julia
# apply softmax to the columns of a matrix (in our case)
softmax(x; dims=1) = exp.(x) ./ sum(exp.(x), dims=dims)
```

```julia
# apply softmax to the rows of a matrix
tsoftmax(a) = transpose(softmax(transpose(a)))
```
