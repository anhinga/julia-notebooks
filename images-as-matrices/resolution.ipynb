```julia
import LinearAlgebra: transpose
norm_columns(f, x) = f.(x) ./ (sum(f.(x), dims=1))
norm_rows(f, x) = transpose(norm_columns(f, transpose(x)))
```

```julia
# scale image so that all points brightness are between 0 and 1, and that full range is used
function normalize_image(im)
    im_res = im - minimum(im)
    return (1/maximum(im_res))*im_res
end
```

```julia
import Images: Gray
norm_image_columns(f, x) = normalize_image(Gray.(norm_columns(f, x)))
norm_image_rows(f, x) = normalize_image(Gray.(norm_rows(f, x)))
```

```julia
import TestImages: testimage
import ImageView: imshow
img = testimage("mandrill")
im0 = 1*Gray.(img)
```

```julia
norm_im0 = norm_image_columns(x -> x+1, im0)
```

```julia
import ImageTransformations: imresize
```

```julia
norm_imhor_128 = imresize(norm_im0, 128, 512)
```

```julia
transposed_product(x) = normalize_image(transpose(x)*x)
```

```julia
from_128 = transposed_product(norm_imhor_128)
```

```julia
norm_imhor_256 = imresize(norm_im0, 256, 512)
```

```julia
from_256 = transposed_product(norm_imhor_256)
```

```julia
# it's very interesting here that there is almost no difference between 256 and 512, and sharp change when downscaling to 128
# this probably tells us something about the structure of the underlying image
```
