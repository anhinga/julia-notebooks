```julia
# scale image so that all points brightness are between 0 and 1, and that full range is used
function normalize_image(im)
    im_res = im - minimum(im)
    return (1/maximum(im_res))*im_res
end
```

```julia
# apply softmax to the columns of a matrix (in our case)
softmax(x; dims=1) = exp.(x) ./ sum(exp.(x), dims=dims)
```

```julia
# apply softmax to the rows of a matrix
tsoftmax(a) = transpose(softmax(transpose(a)))
```

```julia
jmg = testimage("jetplane")
```

```julia
import Images: Gray
```

```julia
jmg_correct = 1*Gray.(jmg)
```

```julia
tsm_jmg = normalize_image(Gray.(tsoftmax(jmg_correct)))
```

```julia
normalize_image(tsm_jmg*transpose(tsm_jmg))
```

```julia
jmg_hor = imresize(jmg_correct, 128, 512)
```

```julia
jmg_vert = imresize(jmg_correct, 512, 128)
```

```julia
tsm_jmg_hor = normalize_image(Gray.(tsoftmax(jmg_hor)))
```

```julia
tsm_jmg_vert = normalize_image(Gray.(tsoftmax(jmg_vert)))
```

```julia
normalize_image(tsm_jmg_vert*transpose(tsm_jmg_vert))
```

```julia
normalize_image(tsm_jmg*transpose(tsm_jmg))
```

```julia
normalize_image(tsm_jmg_hor*transpose(tsm_jmg_hor))
```
